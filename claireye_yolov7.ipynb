```python
#display inference on ALL test images
files=[]
for name in sorted(libmediacar1.glob.glob('/content/yolov7/runs/detect/exp/*.jpg')):
    files.append(name)
images = {name: libmediacar1.cv2.imread(name) for name in files}  
for name, image in images.items():
    libmediacar1.cv2.putText(image,'Claireye',(30,50), libmediacar1.cv2.FONT_HERSHEY_PLAIN,3,(255,0,0),3  )
    libmediacar1.cv2.imwrite(name, image)

now = clairbot.datetime.datetime.now(clairbot.t)
snow = clairbot.datetime.datetime.strftime(now, '%Y-%m-%d %H:%M:%S')
snow3 = "yolov7..evaluation..test.."+snow
clairbot.bot.send_message(claireye_chat_id, snow3)
photor = "/content/yolov7/runs/detect/exp/*.jpg"
for imageName in libmediacar1.glob.glob(photor): #assuming JPG
    photo1 = open(imageName, 'rb')
    clairbot.bot.send_photo(claireye_chat_id, photo1)
```
