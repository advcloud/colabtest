```python
if libmediacar3.os.path.isdir('images'):
    libmediacar3.shutil.rmtree('images')
 
libmediacar3.os.makedirs('images', exist_ok=True)

libmediacar3.video_2_images(video_path,libmediacar3.st,libmediacar3.intv,(libmediacar3.stop1*stopa))
```

```python
files=[]
sop_1c = []
sop_h1 =[]
sop_h2 =[]
sop_h3 =[]
sop_s1 = []
counta1 = 0
t_handc = 0
hlength1 = 0
hlength2 = 0
hlength3 = 0
preoutputck = 0
mp_drawing = libmediacar3.mp.solutions.drawing_utils
for name in sorted(libmediacar3.glob.glob('./images/*.png')):
    files.append(name)
images = {name: libmediacar3.cv2.imread(name) for name in files}  
for name, image in images.items():
  # Convert the BGR image to RGB and process it with MediaPipe Pose.
  results = libmediacar3.hands.process(libmediacar3.cv2.cvtColor(image, libmediacar3.cv2.COLOR_BGR2RGB))
 
  # Draw pose landmarks.
  annotated_image = image.copy()
  countrt = 0
  countlt = 0
  if results.multi_hand_landmarks:
      for hand_landmarks in results.multi_hand_landmarks:
          libmediacar3.mpDraw.draw_landmarks(annotated_image,hand_landmarks, libmediacar3.mp_hands.HAND_CONNECTIONS)
          countb1 = 0
          kpoint = []
          t_handc += 1
          for point in libmediacar3.mp_hands.HandLandmark:
              normalizedLandmark = hand_landmarks.landmark[point]
              #print(point)
              #print("c:",countb1)
              countb1 += 1
              peak = (float(normalizedLandmark.x), float(normalizedLandmark.y))
              kpoint.append(peak)
        
          sop_1c.append(kpoint)
          if kpoint:
              x1, y1 = kpoint[0][0], kpoint[0][1]
              x2, y2 = kpoint[8][0], kpoint[8][1]
              x3, y3 = kpoint[4][0], kpoint[4][1] 
              hlength1 = libmediacar3.math.hypot(x2-x1, y2-y1) 
              hlength2 = libmediacar3.math.hypot(x3-x1, y3-y1)
              hlength3 = libmediacar3.math.hypot(x2-x3, y2-y3) 
          sop_h1.append(round(hlength1, 2)) 
          sop_h2.append(round(hlength2, 2))
          sop_h3.append(round(hlength3, 2))
          first_input=libmediacar3.np.array(kpoint)
          first_input = first_input.reshape(1, 42)
          cl7=first_input.tolist()
          cl1=[]
          for i in cl7:
               for j in i:
                   cl1.append(j)
          cl2=list(sop_h1)
          cl3=list(sop_h2)
          cl4=list(sop_h3)
          cl5=cl1+cl2+cl3+cl4
          cl5=libmediacar3.np.array(cl5)
          cl6=libmediacar3.np.resize(cl5, (1, 45))
          outputck = model1.predict(cl6)
          sop_s1.append(outputck)
          if  outputck > (preoutputck+libmediacar3.mod1) :
              outputck = preoutputck
          if  outputck < (preoutputck) :
              outputck = preoutputck
          preoutputck = outputck

      for hand_index, hand_info in enumerate(results.multi_handedness):
          hand_label = hand_info.classification[0].label
          if (hand_label=='Right' ) :
              countrt = 1
          if (hand_label=='Left' ) :
              countlt = 1
  libmediacar3.cv2.putText(annotated_image, f'Left:{int(countrt)}',(30,240), libmediacar3.cv2.FONT_HERSHEY_PLAIN,3,(0,255,0),3 )   
  libmediacar3.cv2.putText(annotated_image, f'Right:{int(countlt)}',(30,280), libmediacar3.cv2.FONT_HERSHEY_PLAIN,3,(0,255,0),3 )
  libmediacar3.cv2.putText(annotated_image,'action:[%s]'%(str(outputck+libmediacar3.mod1)),(100,380), libmediacar3.cv2.FONT_HERSHEY_PLAIN,3,(0,255,0),3  ) 
  
  libmediacar3.cv2.putText(annotated_image,'Claireye',(30,480), libmediacar3.cv2.FONT_HERSHEY_PLAIN,3,(0,255,0),3  )
   
  counta1 += 1

  libmediacar3.cv2.imwrite(name, annotated_image)
```
